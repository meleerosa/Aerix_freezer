```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from tqdm import tqdm
import os
```

```python
_dir = '_19'
_file_ls = os.listdir(_dir)
_file_ls
```

```python
freez1 = pd.read_csv(os.path.join(_dir, _file_ls[0]))
suth1 = pd.read_csv(os.path.join(_dir, _file_ls[2]))
```

```python
freez1.columns
freez1.columns = ['freezer_id', 'gw_time', 'status', 'fan_run', 'fan_trip',
       'freez_run', 'freez_trip', 'defrost', 'ctrl_mode', 'ctrl_operation',
       'operation', 'run_time', 'insert_dt']
```

```python
suth1.columns
suth1.columns = ['sensor_id', 'gw_time', 'status', 'color_list', 'avg_temperature',
       'avg_humidity', 'temperature', 'humidity', 'battery', 'available',
       'insert_dt']
```

```python
freez1.head(3)
```

```python
suth1.head(3)
```

```python
freezer = freez1.copy()
freezer['key'] = [None]*len(freezer)
```

```python
sensor_id_16 = [431, 440, 449, 461]
freezer_id_16 = [308,314,320,326]
sensor_16 = pd.DataFrame()
freezer_16 = pd.DataFrame()
for elem in sensor_id_16:
    sensor_16 = pd.concat([sensor_16, suth1[suth1['sensor_id'] == elem]])
    print("sensor id {}의 데이터 개수: {}".format(elem, len(suth1[suth1['sensor_id'] == elem])))
for elem in freezer_id_16:
    freezer_16 = pd.concat([freezer_16, freezer[freezer['freezer_id'] == elem]])
    print("freezer id {}의 데이터 개수: {}".format(elem, len(freezer[freezer['freezer_id'] == elem])))


print(sensor_16['sensor_id'].unique())
print(freezer_16['freezer_id'].unique())
```

```python
# 1월 데이터는 29일 부터 수집이므로 freezer 앞부분 데이터 삭제
freezer_16 = freezer_16.sort_values(by = 'gw_time', ascending= True).reset_index(drop = True)
freezer_16 = freezer_16.reset_index(drop = True)
freezer_16
```

```python
sensor_16
```

```python
sensor_16['key'] = [x for x in range(len(sensor_16))]
```

```python
suth_4elem = sensor_16.reset_index(drop = True)
suth_4elem['gw_time'] = pd.to_datetime(suth_4elem['gw_time'])
freezer_16['gw_time'] = pd.to_datetime(freezer_16['gw_time'])
from tqdm import tqdm
for i in tqdm(range(len(suth_4elem))):
    freezer_16.loc[(suth_4elem.loc[i,'gw_time']-freezer_16['gw_time'] <=  dt.timedelta(seconds=30)) & (suth_4elem.loc[i,'gw_time']- freezer_16['gw_time'] >= dt.timedelta(seconds=0)), 'key'] = suth_4elem.loc[i,'key']
```

```python
suth_4elem.sort_values(by = 'gw_time')
suth_4elem['key'] = suth_4elem['key'].astype('object')
```

```python
suth_4elem
```

```python
freezer_16
```

```python
df = pd.merge(freezer_16, suth_4elem, how = 'inner', on = 'key').sort_values(by = 'key')
key = df['key']
gw_time_x = df['gw_time_x']
gw_time_y = df['gw_time_y']
df = df.drop(columns= ['key','gw_time_x','gw_time_y'])
df.insert(0,'key',key)
df.insert(1,'gw_time_x',gw_time_x)
df.insert(2,'gw_time_y',gw_time_y)
df
```
