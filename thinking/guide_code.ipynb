# Data Read

```python
# 1901
freez1 = pd.read_csv(os.path.join(_dir, _file_ls[0]))
suhp1 = pd.read_csv(os.path.join(_dir, _file_ls[1]))
```

```python
print(f"freez sensor: {np.unique(freez1['equip_sensor_id'])}")
print(f"suhp sensor: {np.unique(suhp1['equip_sensor_id'])}")
print(len(np.unique(freez1['equip_sensor_id'])), len(np.unique(suhp1['equip_sensor_id'])))
```

```python
freez1_list = [data for label, data in freez1.groupby(freez1['equip_sensor_id'])]
suhp1_list = [data for label, data in suhp1.groupby(suhp1['equip_sensor_id'])]
```

```python
suhp1_list[0].info()
```

```python
suhp1_list[0].isnull().sum()
```

```python
suhp1_list[0].describe()
```

```python
plt.figure(figsize=(12, 4))
plt.title(f"sensorid: {np.unique(suhp1_list[0]['equip_sensor_id'])}")
plt.plot(pd.to_datetime(suhp1_list[0]['gw_time']), suhp1_list[0]['temperature'])
plt.xlabel('time')
plt.ylabel('temperature')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.title(f"sensorid: {np.unique(suhp1_list[0]['equip_sensor_id'])}")
plt.plot(pd.to_datetime(suhp1_list[0]['gw_time']), suhp1_list[0]['humidity'])
plt.xlabel('time')
plt.ylabel('humidity')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(pd.to_datetime(suhp1_list[0]['gw_time']), suhp1_list[0]['temperature'], label=f"{np.unique(suhp1_list[0]['equip_sensor_id'])}")
plt.plot(pd.to_datetime(suhp1_list[1]['gw_time']), suhp1_list[1]['temperature'], label=f"{np.unique(suhp1_list[1]['equip_sensor_id'])}")
plt.xlabel('time')
plt.ylabel('temperature')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
for i in range(len(suhp1_list)):
    if np.unique(suhp1_list[i]['equip_sensor_id']) not in [[467], [404], [338], [371]]:
        plt.plot(pd.to_datetime(suhp1_list[i]['gw_time']), suhp1_list[i]['temperature'], label=f"{np.unique(suhp1_list[i]['equip_sensor_id'])}")
plt.xlabel('time')
plt.ylabel('temperature')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
for i in range(len(suhp1_list)):
    if np.unique(suhp1_list[i]['equip_sensor_id'])  in [[467], [404], [338], [371]]:
        plt.plot(pd.to_datetime(suhp1_list[i]['gw_time']), suhp1_list[i]['temperature'], label=f"{np.unique(suhp1_list[i]['equip_sensor_id'])}")
plt.xlabel('time')
plt.ylabel('temperature')
plt.legend()
plt.show()
```
